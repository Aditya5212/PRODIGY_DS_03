```python
column = data[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data_encoded = data1.apply(lb.fit_transform)
data_encoded
```

```python
data_encoded['deposit'].value_counts()
```

```python
x = data_encoded.drop('deposit',axis=1)
y = data_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Creating a function to compute Confusion Matrix, Classification Report and to generate training and testing scores

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```

## Plot Decision Tree


```python
from sklearn.tree import plot_tree
```

```python
cn = ['no','yes']
fn = x_train.columns
print(fn)
print(cn)
```
